```python
vernal_equinox = []
summer_solstice = []
autumnal_equinox = []
winter_solstice = []
perihelion = []
aphelion = []

with open('SolarEventsDateTime.txt') as f:
    contents = f.read()

    for line in contents.splitlines():
        if len(line) > 0:
            values = line.split()
            if values[0].isdigit():
                year = values[0]
                vernal_equinox_date = values[1] 
                summer_solstice_date = values[3]
                winter_solstice_date = values[5]
                perihelion_date = values[7]
                aphelion_date = values[9]

                vernal_equinox.append(datetime.strptime(year + '/' + vernal_equinox_date , '%Y/%m/%d').date())
                summer_solstice.append(datetime.strptime(year + '/' + summer_solstice_date , '%Y/%m/%d').date())
                autumnal_equinox.append(datetime.strptime(year + '/' + winter_solstice_date , '%Y/%m/%d').date())
                winter_solstice.append(datetime.strptime(year + '/' + perihelion_date , '%Y/%m/%d').date())
                #perihelion_date.append(datetime.strptime(year + '/' + aphelion_date , '%Y/%m/%d').date())
                aphelion.append(datetime.strptime(year + '/' + aphelion_date , '%Y/%m/%d').date())
```

```python
from harmony import Client
```

```python
# Connect to Earthdata using Harmony client and download files
harmony_client = Client(auth=('CHANGEME', 'CHANGEME'))
base_url = 'https://archive.podaac.earthdata.nasa.gov/podaac-ops-cumulus-protected/MUR-JPL-L4-GLOB-v4.1/'

for day in vernal_equinox:
    filename = day.strftime('%Y%m%d') + '090000-JPL-L4_GHRSST-SSTfnd-MUR-GLOB-v02.0-fv04.1.nc'
    url = base_url + filename
    print(url)
    r = requests.get(url, allow_redirects=True)
    open('downloads/vernal_equinox/' + filename, 'wb').write(r.content)

for day in summer_solstice:
    filename = day.strftime('%Y%m%d') + '090000-JPL-L4_GHRSST-SSTfnd-MUR-GLOB-v02.0-fv04.1.nc'
    url = base_url + filename
    print(url)
    r = requests.get(url, allow_redirects=True)
    open('downloads/summer_solstice/' + filename, 'wb').write(r.content)

for day in autumnal_equinox:
    filename = day.strftime('%Y%m%d') + '090000-JPL-L4_GHRSST-SSTfnd-MUR-GLOB-v02.0-fv04.1.nc'
    url = base_url + filename
    print(url)
    r = requests.get(url, allow_redirects=True)
    open('downloads/autumnal_equinox/' + filename, 'wb').write(r.content)

for day in winter_solstice:
    filename = day.strftime('%Y%m%d') + '090000-JPL-L4_GHRSST-SSTfnd-MUR-GLOB-v02.0-fv04.1.nc'
    url = base_url + filename
    print(url)
    r = requests.get(url, allow_redirects=True)
    open('downloads/winter_solstice/' + filename, 'wb').write(r.content)

for day in perihelion:
    filename = day.strftime('%Y%m%d') + '090000-JPL-L4_GHRSST-SSTfnd-MUR-GLOB-v02.0-fv04.1.nc'
    url = base_url + filename
    print(url)
    r = requests.get(url, allow_redirects=True)
    open('downloads/perihelion/' + filename, 'wb').write(r.content)

for day in aphelion:
    filename = day.strftime('%Y%m%d') + '090000-JPL-L4_GHRSST-SSTfnd-MUR-GLOB-v02.0-fv04.1.nc'
    url = base_url + filename
    print(url)
    r = requests.get(url, allow_redirects=True)
    open('downloads/aphelion/' + filename, 'wb').write(r.content)
```
